STEP 1

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image  # Correct import for PIL
import matplotlib.pylab as plt


```

STEP 2

```python
# List of URLs to fetch images
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://s.kaskus.id/r540x540/images/2019/08/17/10517404_201908171135120907.jpg"
]

# Loop through each URL and process the image
for url in urls:
    # Read the image from the URL
    image = io.imread(url)

    # Resize the image
    image= cv.resize(image, (0, 0), fx=0.5, fy=0.5)

    # Convert the image to RGB format
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    # Concatenate the original and processed images side by side
    final_frame = cv.hconcat([image, image_2])

    # Display the final concatenated image
    cv2_imshow(final_frame)
    print('\n')
```

Step 3
Melihat ukuran file image

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image:tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

Step 4
Mengakses pixel dengan memberi garis diagonal menyilang

```python
# Convert image from BGR to RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)


# Membuat garis horizontal di tengah image
for y in range(lebar):
    image_3[int((tinggi)/2), y] = [255, 255, 255]  # Warna putih (RGB)

# Menggabungkan kedua gambar secara horizontal
final_frame = cv.hconcat([image_2, image_3])

# Menampilkan gambar hasil
cv2_imshow(final_frame)
```

TUGAS

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python


# Garis Vertikal

# Convert image from BGR to RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)



# Membuat garis vertikal di tengah image
for x in range(tinggi):
    image_3[x, int((lebar)/2)] = [255, 0, 0]  # Warna biru (RGB)

# Menggabungkan kedua gambar secara horizontal
final_frame = cv.hconcat([image_2, image_3])

# Menampilkan gambar hasil
cv2_imshow(final_frame)
print('\n')
# Garis Menyilang

# Convert image from BGR to RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis diagonal dari kiri atas ke kanan bawah
for i in range(min(tinggi, lebar)):
    image_3[i, i] = [255, 0, 0]  # Garis diagonal biru

# Membuat garis diagonal dari kanan atas ke kiri bawah
for i in range(min(tinggi, lebar)):
    image_3[i, lebar - i - 1] = [255, 0, 0]  # Garis diagonal biru

# Menggabungkan kedua gambar secara horizontal
final_frame = cv.hconcat([image_2, image_3])

# Menampilkan gambar hasil
cv2_imshow(final_frame)
```

3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang
tertentu

```python
# Convert image from BGR to RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)


# Panjang garis yang diinginkan
panjang_garis = 80

# Posisi awal garis horizontal di tengah gambar
x_start = int((lebar - panjang_garis) / 2)
x_end = x_start + panjang_garis

# Membuat garis horizontal di tengah gambar
image_3[int(tinggi / 2), x_start:x_end] = [255, 0, 0]  # Warna putih (RGB)

# Menggabungkan kedua gambar secara horizontal
final_frame = cv.hconcat([image_2, image_3])

# Menampilkan gambar hasil
cv2_imshow(final_frame)
```

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam
gambar


```python
# Convert image from BGR to RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Dapatkan dimensi gambar
tinggi, lebar, _ = image_rgb.shape

# Tentukan posisi dan ukuran kotak
x_start = 50  # Posisi x awal kotak
y_start = 100  # Posisi y awal kotak
kotak_lebar = 50  # Lebar kotak
kotak_tinggi = 50 # Tinggi kotak

# Tentukan batas kotak
x_end = x_start + kotak_lebar
y_end = y_start + kotak_tinggi

# Gambar kotak putih di gambar
image_rgb[y_start:y_end, x_start:x_end] = [255, 0, 0]  # Warna putih (RGB)

# Menggabungkan kedua gambar secara horizontal
final_frame = cv.hconcat([image_2, image_rgb])

# Menampilkan gambar hasil
cv2_imshow(final_frame)
```
