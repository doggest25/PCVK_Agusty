```python
image = cv.imread('/content/drive/MyDrive/PCVK/Images/activity.jpg')

# Mengubah gambar dari BGR (OpenCV) ke RGB (Matplotlib)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat salinan gambar untuk manipulasi channel
rb_image = image_rgb.copy()  # Untuk Red-Blue
gb_image = image_rgb.copy()  # Untuk Green-Blue

# Menghapus channel yang tidak diinginkan
rb_image[:, :, 1] = 0  # Menghapus channel Green dari gambar RB
gb_image[:, :, 0] = 0  # Menghapus channel Red dari gambar GB

# Menampilkan gambar dengan channel Red-Blue dan Green-Blue
plt.figure(figsize=(10, 5))

# Red-Blue (RB)
plt.subplot(1, 2, 1)
plt.imshow(rb_image)
plt.title('Red-Blue Channel')


# Green-Blue (GB)
plt.subplot(1, 2, 2)
plt.imshow(gb_image)
```

```python
# Membaca gambar asli
image = cv.imread('/content/drive/MyDrive/PCVK/Images/activity.jpg')

# Mengubah gambar dari BGR (OpenCV) ke RGB (Matplotlib)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Memotong gambar pada baris ke-5 hingga ke-30 dan semua kolom
cropped_image = image_rgb[5:30, :]

# Menghapus channel Green dan Blue untuk menampilkan hanya channel Red
red_channel_image = cropped_image.copy()
red_channel_image[:, :, 1] = 0  # Setel channel Green menjadi nol
red_channel_image[:, :, 2] = 0  # Setel channel Blue menjadi nol

# Menampilkan gambar yang sudah dipotong dan hanya channel Red
plt.figure(figsize=(10, 5))
plt.imshow(red_channel_image)
plt.title('Cropped Image (Rows 5-30, All Columns, Red Channel Only)')
plt.axis('off')  # Menyembunyikan sumbu
plt.show()
```
