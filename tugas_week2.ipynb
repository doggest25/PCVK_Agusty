```python
image = cv.imread('/content/drive/MyDrive/PCVK/Images/activity.jpg')

# Mengubah gambar dari BGR (OpenCV) ke RGB (Matplotlib)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat salinan gambar untuk manipulasi channel
rb_image = image_rgb.copy()  # Untuk Red-Blue
gb_image = image_rgb.copy()  # Untuk Green-Blue

# Menghapus channel yang tidak diinginkan
rb_image[:, :, 1] = 0  # Menghapus channel Green dari gambar RB
gb_image[:, :, 0] = 0  # Menghapus channel Red dari gambar GB

# Menampilkan gambar dengan channel Red-Blue dan Green-Blue
plt.figure(figsize=(10, 5))

# Red-Blue (RB)
plt.subplot(1, 2, 1)
plt.imshow(rb_image)
plt.title('Red-Blue Channel')


# Green-Blue (GB)
plt.subplot(1, 2, 2)
plt.imshow(gb_image)
```

```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# Path ke file gambar di Google Drive
image_path = '/content/drive/MyDrive/PCVK/Images/activity.jpg'  # Ubah dengan path gambar Anda

# Buka gambar yang ada
gambar = Image.open(image_path)

# Buat objek untuk menggambar di atas gambar
draw = ImageDraw.Draw(gambar)

# Tentukan lima kotak dengan ukuran dan warna berbeda
kotak = [
    ((50, 50, 150, 150), 'red'),
    ((200, 50, 350, 200), 'green'),
    ((50, 200, 150, 350), 'blue'),
    ((200, 250, 400, 400), 'orange'),
    ((1000, 1750, 500, 2300), 'purple')
]

# Gambar kotak pada gambar
for rect, warna in kotak:
    draw.rectangle(rect, fill=warna)

# Tampilkan gambar menggunakan plt.imshow
plt.imshow(gambar)

```

```python
# Path ke file gambar yang benar
image_path = '/content/drive/MyDrive/PCVK/Images/kitten.jpg'

# Baca gambar dari Google Drive
image = cv.imread(image_path)

# Pastikan gambar terbaca
if image is None:
    print("Gambar tidak ditemukan!")
else:
    # Pisahkan channel B, G, R dari gambar
    B, G, R = cv.split(image)

    # Tentukan ukuran rectangle (misal 50x50 pixel) di sudut kiri bawah
    height, width = B.shape  # Mendapatkan ukuran gambar
    start_point = (0, height - 500)  # Koordinat kiri bawah
    end_point = (500, height)  # Ukuran rectangle 50x50
    color = (255)  # Warna putih (maksimum di channel B)
    thickness = -1  # Isi penuh rectangle

    # Gambar rectangle pada channel B
    cv.rectangle(B, start_point, end_point, color, thickness)

    # Gabungkan kembali channel B, G, dan R
    merged_image = cv.merge([B, G, R])

    # Tambahkan teks pada gambar
    font = cv.FONT_HERSHEY_SIMPLEX  # Pilihan font
    text = "Hello Kitten"
    font_scale = 1.5  # Ukuran font
    font_color = (0, 255, 0)  # Warna font dalam BGR (Hijau)
    font_thickness = 2  # Ketebalan font
    text_position = (50, 100)  # Posisi teks di gambar

    # Tambahkan teks ke gambar
    cv.putText(merged_image, text, text_position, font, font_scale, font_color, font_thickness, cv.LINE_AA)

    # Tampilkan gambar hasil menggunakan matplotlib
    plt.imshow(cv.cvtColor(merged_image, cv.COLOR_BGR2RGB))

    plt.show()
```
