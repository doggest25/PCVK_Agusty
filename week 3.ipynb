2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan
formula yang terdapat pada bagian Ulasan Teori untuk kontras, sehingga
menghasilkan keluaran seperti berikut:


```python
print('Mengubah kontras dan tingkat kecerahan citra')
print('----------------------------------')

try:
    brightness = int(input('Masukkan tingkat kecerahan [-255 - 255]: '))
    kontras = float(input('Masukkan tingkat kontras [1.0 - 3.0]: '))


except ValueError as e:
    print(f'Error: {e}')
    exit()

# Membaca citra
original = cv.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')  # Ganti dengan path ke citra Anda

# Mengatur kontras dan kecerahan
# Kontras dan kecerahan diatur menggunakan cv2.convertScaleAbs
# alpha = kontras (faktor pengali kontras)
# beta = brightness (penambah kecerahan)
adjusted_image = cv.convertScaleAbs(original, alpha=kontras, beta=brightness)

# Menampilkan citra asli dan citra yang disesuaikan
final_frame = cv.hconcat((original, adjusted_image))

# Menampilkan hasil
from google.colab.patches import cv2_imshow
cv2_imshow(final_frame)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory
menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi
log, sehingga menghasilkan keluaran seperti berikut:

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Print header
print('Mengubah tingkat kecerahan citra dengan Transformasi Log')
print('----------------------------------')

# Input brightness value
brightness = float(input('Masukkan nilai kecerahan: '))

# Load the image
image = cv.imread('/content/drive/MyDrive/PCVK/Images/female.tiff')

# Convert the image to float32 for precision
image = image.astype(np.float32)

# Split the image into color channels
channels = cv.split(image)

# Apply the log transformation to each channel
channels = [brightness * np.log1p(channel) for channel in channels]

# Merge the transformed channels back into one image
log_transformed = cv.merge(channels)

# Normalize the result to the range [0, 255]
log_transformed = cv.normalize(log_transformed, None, 0, 255, cv.NORM_MINMAX)

# Convert the image to uint8 for display
log_transformed = log_transformed.astype(np.uint8)

# Display the original and transformed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv.cvtColor(image.astype(np.uint8), cv.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Log Transformed Image')
plt.imshow(cv.cvtColor(log_transformed, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
plt.axis('off')

plt.show()
```

4. Implementasikan transformasi grayscale menggunakan metode averaging, lightness,
dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada
bagian Ulasan Teori, sehingga menghasilkan keluaran seperti berikut:

a. Averaging


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the image
original = cv.imread('/content/drive/MyDrive/PCVK/Images/KTP.jpeg')  # Replace with the path to your image

# Convert BGR to RGB for displaying
image_rgb = cv.cvtColor(original, cv.COLOR_BGR2RGB)

# Convert to grayscale using the average method
# Extract the individual color channels
B, G, R = cv.split(original)

# Calculate the grayscale image using the average formula
gray_avg = (R + G + B) / 3

# Convert the grayscale image to uint8 type
gray_avg = np.uint8(gray_avg)

# Convert grayscale to RGB for consistent display with the original image
gray_rgb = cv.cvtColor(gray_avg, cv.COLOR_GRAY2RGB)

# Display the original and grayscale average images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image_rgb)


# Grayscale Average Image
plt.subplot(1, 2, 2)
plt.title('Grayscale Average')
plt.imshow(gray_rgb)


plt.show()
```

b. Lightness



```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra
original = cv.imread('/content/drive/MyDrive/PCVK/Images/KTP.jpeg')  # Ganti dengan path ke citra Anda

# Convert BGR to RGB for displaying
image_rgb = cv.cvtColor(original, cv.COLOR_BGR2RGB)

# Convert to grayscale using Lightness method
# Split the image into R, G, B channels
r, g, b = cv.split(original)

# Calculate Lightness
lightness = (np.max([r, g, b], axis=0) + np.min([r, g, b], axis=0)) / 2

# Convert lightness to RGB for consistent display with original image
lightness_rgb = cv.cvtColor(lightness.astype(np.uint8), cv.COLOR_GRAY2RGB)

# Display the original and grayscale lightness images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image_rgb)


# Grayscale Lightness Image
plt.subplot(1, 2, 2)
plt.title('Grayscale Lightness')
plt.imshow(lightness_rgb)


plt.show()
```
