```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
from google.colab import files
import cv2 as cv
from google.colab.patches import cv2_imshow  # For image display

# Upload file gambar
uploaded = files.upload()

# Ambil nama file dari upload
filename = list(uploaded.keys())[0]
print(f'Nama file yang di-upload: {filename}')

# Baca gambar
image = cv.imread(filename)

# Cek apakah gambar berhasil dibaca
if image is None:
    print(f'Gagal membaca gambar dari {filename}')
else:
    # Ubah ukuran gambar
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)

    # Tampilkan gambar
    cv2_imshow(image)
```

```python
image = cv.imread('KTP.jpeg')
image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
cv2_imshow(image)

```

```python
image_1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image_1)

```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
tinggi = image.shape[0]
lebar = image.shape[1]
print(f'tinggi: {tinggi}, lebar: {lebar}')
fig, ax = plt.subplots()
rect_kab = plt.Rectangle((int(lebar / 2) + 110, int(tinggi / 2) + 45),
62, 10, linewidth=1, edgecolor='none', facecolor='white')
rect_tgl = plt.Rectangle((int(lebar / 2) + 110, int(tinggi / 2) + 60),
62, 10, linewidth=1, edgecolor='none', facecolor='black')
ax.add_patch(rect_kab)
ax.add_patch(rect_tgl)
plt.imshow(image_2)

```
